### a) check if it is in the alphabet

```python
def is_letter(char):
    '''
    This function chekc the char is in alphabet
        arg:
            char
        return 
            letter : True or False
    '''
    alphabet = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
    letter = False
    if char in alphabet:
        letter = True
 
    return (letter)
```

### b) generator random word.

```python
import random
def random_word(word_list):
    '''
        This function returns a random word from a list
            arg:
                list word
            return
                word
    '''
    random_w =random.choice(word_list)
    return (random_w)
  

words = ["list", "tuple","comprenhensions"]
random_word(words)
```

```python
def pos_char(char,word):
    '''
        This function looks for the character in the word and returns the positions where they appear 
            args :
                char
                word
            return
                positions
    '''
    pos = []
    for i in enumerate(word):
        print(i[1])
        if i[1] == char:
            pos.append(i[0])
        print(pos)
    return (pos)
```
